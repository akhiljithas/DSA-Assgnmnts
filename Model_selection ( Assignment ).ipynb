# 1. Load the dataset into python environment

```python
data = pd.read_csv(r'C:\Users\user\Downloads\titanic_dataset.csv')
```

# 2. Make ‘PassengerId’ as the index column

```python
data.set_index('PassengerId', inplace=True)
```

# 4. Drop the unnecessary columns

```python
data = data.drop(['Name', 'Ticket', 'Cabin'], axis=1)
```

```python
data.columns
```

# 5. Handle the missing values (if any)

```python
data.isna().sum()
```

```python
for column in ['Age', 'Embarked']:
    if column == 'Age':
        data[column] = data[column].fillna(data[column].mean())
    elif column == 'Embarked':
        data[column] = data[column].fillna(data[column].mode()[0])
```

```python
data.isna().sum()
```

# 6. Perform Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
```

```python
columns_to_encode = ['Sex', 'Embarked']
```

```python
data[columns_to_encode] = encoder.fit_transform(data[columns_to_encode])
```

```python
data.head()
```

```python
data.isna().sum()
```

# 7. Split the target feature and perform Scaling

```python
x = data.drop('Survived', axis=1)
y = data['Survived']
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler(feature_range = (0,1))
x = min_max.fit_transform(x)
```

```python
type(x)
```

```python
x = pd.DataFrame(x)
```

```python
x.head()
```

```python
x.columns = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
```

```python
x.head()
```

```python
y.head()
```

```python
x.isna().sum()
```

# 8. kNN with k-fold cross validation

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
```

```python
from sklearn.model_selection import KFold, cross_val_score
kfold_val = KFold(10)
```

```python
for train_index, test_index in kfold_val.split(x,y):
    print('Training Index :', train_index)
    print('Testing Index : ', test_index)
```

```python
kfold_scores_knn = cross_val_score(knn_model, x, y, cv=kfold_val)
kfold_scores_knn
```

```python
average_accuracy_knn = kfold_scores_knn.mean()
average_accuracy_knn 
```

# 9. kNN with Stratified k-fold cross validation

```python
from sklearn.model_selection import StratifiedKFold
skfold_val = StratifiedKFold(n_splits =10)
```

```python
for train_index, test_index in skfold_val.split(x,y):
    print('Training Index :', train_index)
    print('Testing Index : ', test_index)
```

```python
skfold_scores_knn = cross_val_score(knn_model, x, y, cv=skfold_val)
skfold_scores_knn
```

```python
average_accuracy_stratified_knn = skfold_scores_knn.mean()
average_accuracy_stratified_knn
```

# 10. SVM with k-fold cross validation

```python
from sklearn.svm import SVC
svm_model = SVC()
```

```python
kfold_scores_svm = cross_val_score(svm_model, x, y, cv=kfold_val)
kfold_scores_svm
```

```python
average_accuracy_svm = kfold_scores_svm.mean()
average_accuracy_svm
```
